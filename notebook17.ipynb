```python
# importing packages 
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
# loading dataset 
data = sns.load_dataset("tips")
```

```python
# plot the swarmplot 
# hue by size 
# oriented to horizontal 
sns.swarmplot(y = "day", x = "total_bill", hue = "size",  
              orient = "h", data = data)
```

```python
plt.show()
```
